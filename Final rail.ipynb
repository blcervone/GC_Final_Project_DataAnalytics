```python

# Assuming df is your DataFrame
# Selecting relevant columns
relevant_data = df_parq[['actual_arrival_time', 'actual_departure_time']]

# Drop rows with missing values in either of the selected columns
relevant_data = relevant_data.dropna()

# Convert 'actual_arrival_time' and 'actual_departure_time' to datetime if they are not already
relevant_data['actual_arrival_time'] = pd.to_datetime(relevant_data['actual_arrival_time'])
relevant_data['actual_departure_time'] = pd.to_datetime(relevant_data['actual_departure_time'])

# Plot scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(relevant_data['actual_arrival_time'], relevant_data['actual_departure_time'])
plt.title('Scatter Plot of Actual Arrival Time vs Actual Departure Time')
plt.xlabel('Actual Arrival Time')
plt.ylabel('Actual Departure Time')
plt.grid(True)
plt.show()
```
